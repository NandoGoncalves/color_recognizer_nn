```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, BatchNormalization, Conv2D, MaxPool2D, AveragePooling2D, Flatten
from tensorflow.keras import optimizers
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from tensorflow.keras.utils import load_img, img_to_array
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from keras. wrappers.scikit_learn import KerasRegressor
from keras.utils import np_utils

from datetime import datetime

from PIL import Image, ImageEnhance
import cv2
from skimage.io import imread

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

from scipy.stats import reciprocal

import random
import os
```

```python
def load_image(img_path, width = 128, height = 128, show=False, brightness=1, color=1, contrast=1):

    img = load_img(img_path, target_size=(width, height)) #  image.load_img
    
    # https://fr.acervolima.com/amelioration-de-l-image-dans-pil/
    img = ImageEnhance.Brightness(img).enhance(brightness)
    img = ImageEnhance.Color(img).enhance(color)
    img = ImageEnhance.Contrast(img).enhance(contrast)
    
    img_tensor = img_to_array(img)                    # image. (height, width, channels)
    img_tensor = np.expand_dims(img_tensor, axis=0)         # (1, height, width, channels), add a dimension because the model expects this shape: (batch_size, height, width, channels)
    img_tensor /= 255.                                      # imshow expects values in the range [0, 1]

    if show:
        plt.imshow(img_tensor[0])                           
        plt.axis('off')
        plt.show()

    return img_tensor
```

```python
def build_model(n_conv2D = 2, neurons = 10, filters = 20, kernel_size = 4, strides = 2, padding = "valid", 
                activation='relu', data_format = 'channels_last', pool_size = 4, rate = 0.2, 
                learning_rate = 0.0001, input_shape=(128, 128, 3)):
    
    cnn_model = Sequential()
    
    for c2D in range(n_conv2D):
        cnn_model.add(
            Conv2D(
                input_shape=input_shape, 
                filters=filters, 
                kernel_size=kernel_size, 
                strides=strides, 
                padding=padding,
                activation=activation,
                data_format=data_format
            )
        )
        
        cnn_model.add(MaxPool2D(pool_size=pool_size, data_format=data_format))
        
    cnn_model.add(Dropout(rate))
    cnn_model.add(Flatten())
    
    cnn_model.add(Dense(neurons, activation="softmax"))
    
    cnn_model.compile(
    optimizer=optimizers.Adam(learning_rate=learning_rate), 
    loss='categorical_crossentropy', # binary_crossentropy  
    metrics=['accuracy'])
    
    return cnn_model
    
```

```python
root = r'./dataset/images'

X = []
y = []
labels = {}
label_index = -1

images_per_label = -1
images_per_label_max = 300

for path, subdirs, files in os.walk(root):
    for d in subdirs:
        
        images_per_label = 0
        
        label_index = label_index + 1
        labels[label_index] = d # len(subdirs) - (label_index+1)
        
        d = os.path.join(root, d)
        
        print(d , '=========================================================================')
        
        list_dir = os.listdir(d)
        for i in range(0, images_per_label_max):
            
            filename = list_dir[random.randint(0, len(list_dir)-1)]
            f = os.path.join(d, filename)
            
            X.insert(len(X),cv2.cvtColor(cv2.imread(f), cv2.COLOR_BGR2RGB))
            y.insert(len(y),label_index)

            images_per_label = images_per_label + 1

            print(d, label_index, f)            
        
```

```python
labels
```

```python
X = np.array(X)
y = np.array(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
cnn_model = build_model()
```

```python
cnn_reg = KerasRegressor(build_model)
```

```python
cnn_reg.fit(X_train, y_train, epochs=100, validation_split = 0.3, callbacks=EarlyStopping(patience=10))
```

```python
mse_test = cnn_reg.score(X_test, np_utils.to_categorical(y_test))
```

```python
rnd_search_cv = RandomizedSearchCV(cnn_reg, param_distribs, n_iter=10, cv = 8)
```

```python
rnd_search_cv.fit(X_train, np_utils.to_categorical(y_train), epochs=100, validation_split = 0.3, callbacks=EarlyStopping(patience=10))
```

```python
input_shape=(128, 128, 3)
target_size=(128,128)
batch_size = 16
epochs = 10
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.25 ) # shear_range=45.0,   brightness_range=(0.1, 0.9)
```

```python
train_generator = train_datagen.flow_from_directory(
    './dataset/images', 
    color_mode='rgb',
    target_size=target_size,
    shuffle=True,
    batch_size=batch_size,
    class_mode='categorical',
    subset="training"
)

validation_generator = train_datagen.flow_from_directory(
    './dataset/images', # same directory as training data
    target_size=target_size,
    batch_size=batch_size,
    shuffle=True,
    class_mode='categorical',
    color_mode='rgb',subset='validation') #  set as validation data

es = [EarlyStopping(
    monitor='loss', 
    mode='auto', 
    min_delta=0.0001, 
    patience=10, 
    verbose=1)]
```

```python
training_history = cnn_model.fit(
    train_generator, 
    steps_per_epoch= train_generator.samples // batch_size, # batch_size, # 
    epochs=epochs, 
    
    validation_data = validation_generator, 
    validation_steps = validation_generator.samples // batch_size, # batch_size, #   
    
    verbose=1) #  , callbacks=es
```

```python
# Labels des axes
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

# Courbe de la précision sur l'échantillon d'entrainement
plt.plot(np.arange(1 , len(training_history.history['accuracy'])+1, 1),
         training_history.history['accuracy'],
         label = 'Training Accuracy',
         color = 'blue')


# Courbe de la précision sur l'échantillon de validation
plt.plot(np.arange(1 , len(training_history.history['val_accuracy'])+1, 1),
         training_history.history['val_accuracy'], 
         label = 'Validation Accuracy',
         color = 'red')


# Affichage de la légende
plt.legend()

# Affichage de la figure
plt.show()
```

```python
Y_pred = cnn_model.predict(validation_generator, batch_size)
y_pred = np.argmax(Y_pred, axis=1)

print(len(validation_generator.classes), len(y_pred))

cm = confusion_matrix(validation_generator.classes, y_pred)
print(cm)
```

```python
labels = validation_generator.class_indices.keys()
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)

disp.plot(cmap=plt.cm.Reds)

fig = disp.ax_.get_figure() 
fig.set_figwidth(10)
fig.set_figheight(10) 

plt.show()
```
