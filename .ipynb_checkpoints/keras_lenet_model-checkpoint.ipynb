```python
from keras.datasets.mnist import load_data
import numpy as np
import os
import cv2
import random

from sklearn.model_selection import train_test_split
from sklearn import metrics
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils

import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
#export CUDA_VISIBLE_DEVICES=1
import tensorflow as tf
#gpu_options = tf.GPUOptions(per_process_gpu_memory_fraction=0.8) #0.333
#sess = tf.Session(config=tf.ConfigProto(log_device_placement=True, gpu_options=gpu_options))
tf.compat.v1.GPUOptions(per_process_gpu_memory_fraction=0.7)

from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
```

```python
root = r'./dataset/images'

X = []
y = []
labels = {}
label_index = -1

images_per_label = -1
images_per_label_max = 100


for path, subdirs, files in os.walk(root):
    for d in subdirs:
        
        images_per_label = 0
        
        label_index = label_index + 1
        labels[label_index] = d
        
        d = os.path.join(root, d)
        
        print(d , '=========================================================================')
        
        for filename in os.listdir(d):
            
            f = os.path.join(d, filename)
            
            # chantillonnage les images
            if random.randint(1, 3) in [1]:
                

                X.append(cv2.imread(f))
                y.append(label_index)

                images_per_label = images_per_label + 1
            
            if images_per_label >= images_per_label_max:
                break
            
            print(f)
```

```python
labels
```

```python
X = np.array(X)
y = np.array(y)

print(type(X), type(y))
```

```python
len(X[0][0])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Diviser les pixels des donnes X_train et X_test par 255 afin qu'ils soient compris entre 0 et 1.

Transformer les labels de y_train et y_test en vecteurs catgorielles binaires (one hot), grce  la fonction to_categorical du sous-module np_utils de keras.

```python
X_train = X_train / 255
X_test = X_test / 255

y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

L'architecture du LeNet est constitue des couches suivantes:
<ul>
<li>Convolution 1 : 30 filtres, dimension d'entre (128, 128, 1), dimension du noyau (5, 5), fonction d'activation ReLU, pas de dpassement du noyau.
<li>Max-Pooling 1 : dimension du pooling (2, 2).

<li>Convolution 2 : 16 filtres, dimension du noyau (3, 3), fonction d'activation ReLU, Pas de dpassement du noyau.

<li>Max-Pooling 2 : dimension du pooling (2, 2).

<li>Dropout : Connexions coupes: 20%.

<li>Aplatissement
<li>Dense 1 : 128 neurones, fonction d'activation ReLU.
<li>Dense 2 : 10 neurones, fonction d'activation softmax.
</ul>

## LeNet

```python
batch_size = 16
epochs = 16
```

```python
# Architecture du modle
lenet = Sequential()

conv_1 = Conv2D(filters = 30,                     # Nombre de filtres
                kernel_size = (5, 5),            # Dimensions du noyau
                padding = 'valid',               # Mode de Dpassement
                input_shape = (128, 128, 3),       # Dimensions de l'image en entre
                activation = 'relu')             # Fonction d'activation

max_pool_1 = MaxPooling2D(pool_size = (2, 2))

conv_2 = Conv2D(filters = 16,                    
                kernel_size = (3, 3),          
                padding = 'valid',             
                activation = 'relu')

max_pool_2 = MaxPooling2D(pool_size = (2, 2))

flatten = Flatten()

dropout = Dropout(rate = 0.2)

dense_1 = Dense(units = 128,
                activation = 'relu')

dense_2 = Dense(units = 10,
                activation = 'softmax')

lenet.add(conv_1)
lenet.add(max_pool_1)
lenet.add(conv_2)
lenet.add(max_pool_2)

lenet.add(dropout)
lenet.add(flatten)
lenet.add(dense_1)
lenet.add(dense_2)

# Compilation
lenet.compile(loss='categorical_crossentropy',  # fonction de perte
              optimizer='adam',                 # algorithme de descente de gradient
              metrics=['accuracy'])             # mtrique d'valuation

# Entranement_1
training_history_lenet = lenet.fit(X_train, y_train,          # donnes
                                   validation_split = 0.3,    # split de validation
                                   epochs = epochs,               # nombre d'epochs
                                   batch_size = batch_size)          # taille des batchs
```

```python
train_acc_lenet = training_history_lenet.history['accuracy']
val_acc_lenet = training_history_lenet.history['val_accuracy']
```

## Rseau Dense

```python
dense = Sequential()
dense_0 = Flatten()
dense_1 = Dense(units = 20, input_shape=(128,128,3), kernel_initializer ='normal', activation ='tanh') # input_dim = 49152            
dense_2 = Dense(units = 10, kernel_initializer ='normal', activation ='softmax')
dense.add(dense_0)
dense.add(dense_1)
dense.add(dense_2)
```

```python
dense.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])   
```

```python
training_history_dense = dense.fit(X_train, y_train, validation_split = 0.2, epochs = epochs, batch_size = batch_size, verbose = 1)
```

```python
val_acc_dense = training_history_dense.history['val_accuracy']
```

## CNN

```python
cnn = Sequential()
cnn_1 = Conv2D(filters = 32, kernel_size = (5, 5), padding = 'valid', input_shape = (128, 128, 3), activation = 'relu')
cnn_2 = MaxPooling2D(pool_size = (2, 2))
cnn_3 = Dropout(rate = 0.2)
cnn_4 = Flatten()
cnn_5 = Dense(units = 128, activation = 'relu')
cnn_6 = Dense(units = 10, activation='softmax')

cnn.add(cnn_1)
cnn.add(cnn_2)
cnn.add(cnn_3)
cnn.add(cnn_4)
cnn.add(cnn_5)
cnn.add(cnn_6)
```

```python
cnn.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])  
```

```python
training_history_cnn = cnn.fit(X_train, y_train, validation_split = 0.2, epochs = 16, batch_size = 200, verbose = 1)
```

```python
val_acc_cnn = training_history_cnn.history['val_accuracy']
```

#### Evolutions des prcisions

```python
# Labels des axes
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

# Courbe du score de validation du rseau Dense
plt.plot(np.arange(1 , 17, 1),
         val_acc_dense,
         label = 'Dense',
         color = 'blue')

# Courbe du score de validation du rseau LeNet
plt.plot(np.arange(1 , epochs+1, 1),
         val_acc_lenet, 
         label = 'LeNet',
         color = 'green')

# Courbe du score de validation du rseau CNN
plt.plot(np.arange(1 , 17, 1),
         val_acc_cnn, 
         label = 'CNN',
         color = 'red')

# Affichage de la lgende
plt.legend()

# Affichage de la figure
plt.show()
```

### Prdiction des classes de l'chantillon X_test

```python
test_pred_lenet = lenet.predict(X_test)
test_pred_dense = dense.predict(X_test)
test_pred_cnn = cnn.predict(X_test)

test_pred_lenet_class = test_pred_lenet.argmax(axis = 1)
test_pred_dense_class = test_pred_dense.argmax(axis = 1)
test_pred_cnn_class = test_pred_cnn.argmax(axis = 1)
y_test_class = y_test.argmax(axis = 1)
```

### Compte-rendu valuatif dtaill de la perfomance du modle lenet

```python
print(metrics.classification_report(y_test_class, test_pred_lenet_class))
```

#### voir les images de la base de test o les trois modles se sont tromps

```python
error_indexes = []
for i in range(len(test_pred_cnn)):
    if (test_pred_lenet_class[i] != y_test_class[i]):
        if(test_pred_dense_class[i] != y_test_class[i]):
            if(test_pred_cnn_class[i] != y_test_class[i]):
                error_indexes += [i]

j = 1
for i in np.random.choice(error_indexes, size = 3):
    img = X_test[i] 
    img = img.reshape(128, 128, 3)
    
    plt.subplot(1, 3, j)
    j = j + 1
    plt.axis('off')
    plt.imshow(img,cmap = cm.binary, interpolation='None')
    plt.title('True Label: ' + str(y_test_class[i]) \
              + '\n' + 'Prediction: '+ str(test_pred_lenet_class[i]) \
              + '\n' + 'Confidence: '+ str(round(test_pred_lenet[i][test_pred_lenet_class[i]], 2)))
```
